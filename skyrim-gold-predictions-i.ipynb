```python
import pandas as pd

# Gathers data from the file for the dataset
skyrim_file_path = '../input/skyrim-weapons-dataset/Skyrim_Weapons.csv'
skyrim_data = pd.read_csv(skyrim_file_path)
skyrim_data.columns

# Drops the speed number since most of the speed stats on weapons are null anyway.
skyrim_data = skyrim_data.dropna(axis=0)
print(skyrim_data)

# Selecting a prediction target (damage)
y = skyrim_data.Gold

weapon_features = ['Name', 'Damage', 'Upgrade', 'Type', 'Gold']
X = skyrim_data[weapon_features]

X.describe()
X.head()
X = X.drop(columns=['Name'])
```

```python
from sklearn.tree import DecisionTreeRegressor

X_encoded = pd.get_dummies(X)
skyrim_model = DecisionTreeRegressor(random_state=1)
skyrim_model.fit(X_encoded, y)

print("Making predictions for the following 5 Weapons:")
print(X_encoded.head())
print("The predictions are")
print(skyrim_model.predict(X_encoded.head()))
```
